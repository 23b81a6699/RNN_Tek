```python
import numpy as np 
import pandas as pandas
import tensorflow as tf
import matplotlib.pyplot as plt 

from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Embedding , SimpleRNN , Dense
```

```python
# Load dataset

(x_train , y_train) , (x_test , y_test) = imdb.load_data(num_words=10000)
```

```python
# Dataset shape

print(x_train.shape)
print(x_test.shape)

print(y_train.shape)
print(y_test.shape)
```

```python
# View first review

print(x_train[0])
```

```python
# View first label

print(y_train[0])
```

1 ---> positive review

0 ---> negative review

```python
# Pad sequences

x_train = pad_sequences(
    x_train,
    maxlen=500
)

x_test = pad_sequences(
    x_test,
    maxlen=500
)
```

```python
# After padding

print(x_train.shape)
print(x_test.shape)
```

```python
# Build RNN Model

# Create model

model = Sequential()
```

```python
# Embedding layer

model.add(
    Embedding(
        input_dim=10000,
        output_dim=32,
        input_length=500
    )
)
```

```python
# Simple RNN layer

model.add(
    SimpleRNN(
        32,
        activation='relu'
    )
)
```

```python
# Output layer

model.add(
    Dense(
        1,
        activation='sigmoid'
    )
)
```

```python
# model.build(input_shape=(None, 500))
```

```python
# Model summary

model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    x_train,
    y_train,
    epochs=5,
    batch_size=64,
    validation_split = 0.2
)
```

```python
loss , acc = model.evaluate(x_test , y_test)

print(acc)
```

```python
model.save("../models/imdb_lstm.keras")
```
